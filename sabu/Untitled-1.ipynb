```python


f1 = df['f1']
f2 = df['f2']
f3 = df['f3']
f4 = df['f4']

breath = 45
length = 60

timestamps = []
cop_x_values = []
cop_y_values = []

for i in range(len(df)):
    row = df.loc[i]

    w = row['f1'] + row['f2'] + row['f3'] + row['f4']

    if w >= 10:
        cop_x = breath * ((row['f1'] + row['f2']) - (row['f3'] + row['f4'])) / (2 * w)
        timestamps.append(row['timestamp'])
        cop_x_values.append(cop_x)

        cop_y = length * ((row['f2'] + row['f3']) - (row['f1'] + row['f4'])) / (2 * w)
        cop_y_values.append(cop_y)

    else:
        cop_x = np.nan
        cop_y = np.nan
        timestamps.append(row['timestamp'])
        cop_x_values.append(cop_x)
        cop_y_values.append(cop_y)



plt.plot(cop_x_values,cop_y_values)
plt.xlabel('COP X-coordinate')
plt.ylabel('COP Y-coordinate')
plt.title('Center of Pressure (COP)')
plt.legend()
plt.show()
```
